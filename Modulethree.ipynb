We consider a lattice of sites:

$$
i = 1, 2, ..., N
$$

Each site at time 'n' has a state:

$$
\theta_n(i)
$$

which evolves in time due to:
- Local nonlinear dynamics
- Coupling with neighboring sites

# Governing Equation

The govening equation here is given as:

$$
\theta_{n+1}(i) = (1 - \epsilon)\, f(\theta_n(i)) + \frac{\epsilon}{2} \left[ f(\theta_n(i+1)) + f(\theta_n(i-1)) \right]
$$

### Parameters:
- $f(\theta)$ :map  
- $\epsilon$ : coupling strength  
- $N$ : number of lattice sites 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# BASIC CODE
#First of all I have written the basic code here for plotting the CML for the given equation.

# parameters
N = 100          # number of sites
T = 100          # time steps
K=1 # Non linearity parameter (Fixed)
epsilon = 0.5 #coupling strength (to be varied later in next cell)
omega = 0.1 #omega  (to be varied later in next cell)



theta = np.random.rand(N)
data = np.zeros((T, N))

def f(theta):
    return theta + omega - ((K / (2 * np.pi))) * np.sin(2 * np.pi * theta)

# time evolution
for t in range(T):
    theta_new = np.zeros(N)
    for i in range(N):
        left = theta[(i-1)%N]
        right = theta[(i+1)%N]
        theta_new[i] = ((1 - epsilon)*f(theta[i]) + (epsilon/2)*(f(left) + f(right)) )%1
    theta = theta_new
    data[t] = theta

# plot
plt.figure(figsize=(8,6))
plt.imshow(data, aspect='auto')
plt.colorbar()
plt.xlabel("site i")
plt.ylabel("Time n")
plt.title("CML for the given equation")
plt.show()
```

Interpretation:

  - (i)That is for $$\epsilon=0.1 , \omega=0.1$$
  - (ii)Same is for $$\epsilon =0.1 , \omega = 0.9$$
  Only difference is that both the maps are complementary in the sense that the phase tends to minimum in first case whereas it tends to maximum in the second case.

     $$\epsilon=0.5,0.9,1$$  $$\omega=0.5$$
